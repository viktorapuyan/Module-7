#Exercise 2

* With faang, use type converstion to change the date column into a datetime and the volume column into integers. Then, sort by date and ticker.




```python
faang = pd.read_csv('/content/faang.csv')

faang.dtypes
```

```python
import pandas as pd

faang.loc[:, 'date'] = pd.to_datetime(faang.date) # changing the dtype of date into datetime
faang.astype({'volume': int}) # changing the volume dtype into integer
faang.dtypes
```

```python
faang.sort_values(by=['date','ticker'], inplace=True) #sorting by date and ticker
faang
```

* Find the seven rows with the highest value for volume.

```python
faang_sortvol = faang.sort_values(by=['volume'], ascending=False).head(7)
faang_sortvol
```

* Right now, the data is somewhere between long and wide format. Use melt() to make it comepletely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have seperate columns for open, high, low, close, and volume.

```python
faang_melt = faang.melt( id_vars=['date', 'ticker'],
                               value_vars=['open','high','low','close', 'volume'])
faang_melt
```
